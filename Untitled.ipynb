```python
import pandas as pd
import random
from datetime import datetime, timedelta

# -----------------------------
# Configuration
# -----------------------------
NUM_RECORDS = 1000

districts = [
    "Kigali City", "Huye", "Musanze", "Rubavu", "Rwamagana",
    "Nyagatare", "Muhanga", "Gicumbi", "Rusizi", "Karongi"
]

mental_conditions = [
    "Depression",
    "Anxiety Disorder",
    "PTSD",
    "Bipolar Disorder",
    "Substance Abuse",
    "Schizophrenia"
]

severity_levels = ["Mild", "Moderate", "Severe"]

treatments = [
    "Counseling",
    "Medication",
    "Psychotherapy",
    "Rehabilitation",
    "Combined Therapy"
]

outcomes = ["Improved", "Stable", "Deteriorated"]

genders = ["Male", "Female"]

# -----------------------------
# Helper function for random date
# -----------------------------
def random_date(start, end):
    delta = end - start
    random_days = random.randint(0, delta.days)
    return start + timedelta(days=random_days)

# -----------------------------
# Generate dataset
# -----------------------------
data = []

for i in range(1, NUM_RECORDS + 1):
    patient_id = f"PAT{i:05d}"
    age = random.randint(15, 80)
    gender = random.choice(genders)
    district = random.choice(districts)
    condition = random.choice(mental_conditions)
    severity = random.choices(severity_levels, weights=[0.4, 0.35, 0.25])[0]
    treatment = random.choice(treatments)
    outcome = random.choices(outcomes, weights=[0.6, 0.25, 0.15])[0]

    visit_date = random_date(
        datetime(2022, 1, 1),
        datetime(2024, 12, 31)
    )

    data.append([
        patient_id,
        age,
        gender,
        district,
        condition,
        severity,
        treatment,
        outcome,
        visit_date
    ])

# -----------------------------
# Create DataFrame
# -----------------------------
columns = [
    "Patient_ID",
    "Age",
    "Gender",
    "District",
    "Mental_Health_Condition",
    "Severity_Level",
    "Treatment_Type",
    "Outcome",
    "Visit_Date"
]

df = pd.DataFrame(data, columns=columns)

# -----------------------------
# Save dataset
# -----------------------------
df.to_csv("rwanda_mental_health_hospital_data.csv", index=False)

print("Dataset generated successfully!")
print(df.head())
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS Fact_Mental_Health_Visit (
    Visit_ID INT AUTO_INCREMENT PRIMARY KEY,
    Patient_ID VARCHAR(10),
    Time_ID INT,
    District_ID INT,
    Condition_ID INT,
    Treatment_ID INT,
    Severity_Level VARCHAR(20),
    Outcome VARCHAR(20),
    Visit_Count INT DEFAULT 1,
    FOREIGN KEY (Patient_ID) REFERENCES Dim_Patient(Patient_ID),
    FOREIGN KEY (Time_ID) REFERENCES Dim_Time(Time_ID),
    FOREIGN KEY (District_ID) REFERENCES Dim_District(District_ID),
    FOREIGN KEY (Condition_ID) REFERENCES Dim_Condition(Condition_ID),
    FOREIGN KEY (Treatment_ID) REFERENCES Dim_Treatment(Treatment_ID)
)
""")
conn.commit()
print("Fact table created successfully ")
```

```python
# -----------------------------
# Mental Health DW BI Prep Script
# -----------------------------

import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import os

# 1 Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",  # <-- Replace with your MySQL password
    database="mental_health_dw"
)
cursor = conn.cursor()

# Create folder to save BI CSVs
output_folder = "BI_Exports"
os.makedirs(output_folder, exist_ok=True)

# -----------------------------
# 2 OLAP Queries & Export
# -----------------------------

# 2.1 Visits by Condition
query_condition = """
SELECT c.Condition_Name, COUNT(f.Visit_ID) AS Total_Visits
FROM Fact_Mental_Health_Visit f
JOIN Dim_Condition c ON f.Condition_ID = c.Condition_ID
GROUP BY c.Condition_Name
ORDER BY Total_Visits DESC;
"""
df_condition = pd.read_sql(query_condition, conn)
df_condition.to_csv(os.path.join(output_folder, "Visits_by_Condition.csv"), index=False)
df_condition.plot(kind='bar', x='Condition_Name', y='Total_Visits', title='Visits by Condition')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "Visits_by_Condition.png"))
plt.close()

# 2.2 Visits by District
query_district = """
SELECT d.District_Name, COUNT(f.Visit_ID) AS Total_Visits
FROM Fact_Mental_Health_Visit f
JOIN Dim_District d ON f.District_ID = d.District_ID
GROUP BY d.District_Name
ORDER BY Total_Visits DESC;
"""
df_district = pd.read_sql(query_district, conn)
df_district.to_csv(os.path.join(output_folder, "Visits_by_District.csv"), index=False)
df_district.plot(kind='bar', x='District_Name', y='Total_Visits', title='Visits by District')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "Visits_by_District.png"))
plt.close()

# 2.3 Severity Distribution
query_severity = """
SELECT Severity_Level, COUNT(*) AS Total_Visits
FROM Fact_Mental_Health_Visit
GROUP BY Severity_Level
ORDER BY FIELD(Severity_Level, 'Severe', 'Moderate', 'Mild');
"""
df_severity = pd.read_sql(query_severity, conn)
df_severity.to_csv(os.path.join(output_folder, "Severity_Distribution.csv"), index=False)
df_severity.plot(kind='pie', y='Total_Visits', labels=df_severity['Severity_Level'], autopct='%1.1f%%', title='Severity Distribution')
plt.ylabel('')
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "Severity_Distribution.png"))
plt.close()

# 2.4 Monthly Trend of Visits
query_trend = """
SELECT CONCAT(t.Month, '-', t.Year) AS Month_Year, COUNT(f.Visit_ID) AS Total_Visits
FROM Fact_Mental_Health_Visit f
JOIN Dim_Time t ON f.Time_ID = t.Time_ID
GROUP BY t.Year, t.Month
ORDER BY t.Year, MONTH(STR_TO_DATE(t.Month,'%M'));
"""
df_trend = pd.read_sql(query_trend, conn)
df_trend.to_csv(os.path.join(output_folder, "Monthly_Trend.csv"), index=False)
df_trend.plot(kind='line', x='Month_Year', y='Total_Visits', title='Monthly Trend of Visits')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "Monthly_Trend.png"))
plt.close()

# 2.5 Treatment Outcomes
query_treatment = """
SELECT tr.Treatment_Type, f.Outcome, COUNT(*) AS Count
FROM Fact_Mental_Health_Visit f
JOIN Dim_Treatment tr ON f.Treatment_ID = tr.Treatment_ID
GROUP BY tr.Treatment_Type, f.Outcome
ORDER BY tr.Treatment_Type;
"""
df_treatment = pd.read_sql(query_treatment, conn)
df_treatment.to_csv(os.path.join(output_folder, "Treatment_Outcomes.csv"), index=False)
df_treatment_pivot = df_treatment.pivot(index='Treatment_Type', columns='Outcome', values='Count').fillna(0)
df_treatment_pivot.plot(kind='bar', stacked=True, title='Treatment Outcomes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "Treatment_Outcomes.png"))
plt.close()

# -----------------------------
# 3 Finish
# -----------------------------
conn.close()
print(f" BI data prepared and charts saved in folder: {output_folder}")
```

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Connect to DW
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="mental_health_dw"
)

# Load monthly visits
query = """
SELECT CONCAT(t.Year, '-', LPAD(MONTH(STR_TO_DATE(t.Month,'%M')),2,'0')) AS Month_Year,
       COUNT(f.Visit_ID) AS Total_Visits
FROM Fact_Mental_Health_Visit f
JOIN Dim_Time t ON f.Time_ID = t.Time_ID
GROUP BY t.Year, t.Month
ORDER BY t.Year, MONTH(STR_TO_DATE(t.Month,'%M'));
"""
df = pd.read_sql(query, conn)
conn.close()

# Convert to datetime
df['Month_Year'] = pd.to_datetime(df['Month_Year'])
df.set_index('Month_Year', inplace=False)  # keep as column for Seaborn

# Plot historical trend
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x='Month_Year', y='Total_Visits')  # explicitly set x and y
plt.title("Historical Monthly Mental Health Visits")
plt.ylabel("Total Visits")
plt.xlabel("Month-Year")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Forecast Future Visits (Time Serie

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import pandas as pd

# Make sure Month_Year is the index as datetime
df['Month_Year'] = pd.to_datetime(df['Month_Year'])
df.set_index('Month_Year', inplace=True)

# Fit ARIMA model
model = ARIMA(df['Total_Visits'], order=(1,1,1))
model_fit = model.fit()

# Forecast next 12 months
forecast = model_fit.forecast(steps=12)

# Correctly generate future dates
last_date = df.index[-1]  # Timestamp
forecast_dates = pd.date_range(start=last_date + pd.DateOffset(months=1), periods=12, freq='MS')

# Plot historical + forecast
plt.figure(figsize=(10,5))
plt.plot(df.index, df['Total_Visits'], label='Historical')
plt.plot(forecast_dates, forecast, label='Forecast', color='red')
plt.title("Forecast of Monthly Mental Health Visits")
plt.ylabel("Total Visits")
plt.xlabel("Month-Year")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# -----------------------------
# Mental Health DW: BI + AI Workflow
# -----------------------------
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
import os

# -----------------------------
# 1 Connect to MySQL DW
# -----------------------------
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",  # <-- Replace with your password
    database="mental_health_dw"
)

# Create folder to save BI outputs
output_folder = "BI_AI_Exports"
os.makedirs(output_folder, exist_ok=True)

# -----------------------------
# 2 OLAP Queries & Charts
# -----------------------------

# Visits by Condition
query_condition = """
SELECT c.Condition_Name, COUNT(f.Visit_ID) AS Total_Visits
FROM Fact_Mental_Health_Visit f
JOIN Dim_Condition c ON f.Condition_ID = c.Condition_ID
GROUP BY c.Condition_Name
ORDER BY Total_Visits DESC;
"""
df_condition = pd.read_sql(query_condition, conn)
df_condition.to_csv(os.path.join(output_folder, "Visits_by_Condition.csv"), index=False)

plt.figure(figsize=(10,5))
sns.barplot(data=df_condition, x='Condition_Name', y='Total_Visits', palette='viridis')
plt.xticks(rotation=45)
plt.title("Visits by Condition")
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "Visits_by_Condition.png"))
plt.close()

# Visits by District
query_district = """
SELECT d.District_Name, COUNT(f.Visit_ID) AS Total_Visits
FROM Fact_Mental_Health_Visit f
JOIN Dim_District d ON f.District_ID = d.District_ID
GROUP BY d.District_Name
ORDER BY Total_Visits DESC;
"""
df_district = pd.read_sql(query_district, conn)
df_district.to_csv(os.path.join(output_folder, "Visits_by_District.csv"), index=False)

plt.figure(figsize=(10,5))
sns.barplot(data=df_district, x='District_Name', y='Total_Visits', palette='magma')
plt.xticks(rotation=45)
plt.title("Visits by District")
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "Visits_by_District.png"))
plt.close()

# Severity Distribution
query_severity = """
SELECT Severity_Level, COUNT(*) AS Total_Visits
FROM Fact_Mental_Health_Visit
GROUP BY Severity_Level
ORDER BY FIELD(Severity_Level, 'Severe', 'Moderate', 'Mild');
"""
df_severity = pd.read_sql(query_severity, conn)
df_severity.to_csv(os.path.join(output_folder, "Severity_Distribution.csv"), index=False)

plt.figure(figsize=(6,6))
plt.pie(df_severity['Total_Visits'], labels=df_severity['Severity_Level'], autopct='%1.1f%%', colors=['red','orange','yellow'])
plt.title("Severity Distribution")
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "Severity_Distribution.png"))
plt.close()

# Treatment Outcomes
query_treatment = """
SELECT tr.Treatment_Type, f.Outcome, COUNT(*) AS Count
FROM Fact_Mental_Health_Visit f
JOIN Dim_Treatment tr ON f.Treatment_ID = tr.Treatment_ID
GROUP BY tr.Treatment_Type, f.Outcome
ORDER BY tr.Treatment_Type;
"""
df_treatment = pd.read_sql(query_treatment, conn)
df_treatment.to_csv(os.path.join(output_folder, "Treatment_Outcomes.csv"), index=False)

df_treatment_pivot = df_treatment.pivot(index='Treatment_Type', columns='Outcome', values='Count').fillna(0)
df_treatment_pivot.plot(kind='bar', stacked=True, figsize=(10,5), colormap='Set2')
plt.title("Treatment Outcomes")
plt.ylabel("Number of Visits")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "Treatment_Outcomes.png"))
plt.close()

# Monthly Trend
query_trend = """
SELECT CONCAT(t.Year, '-', LPAD(MONTH(STR_TO_DATE(t.Month,'%M')),2,'0')) AS Month_Year,
       COUNT(f.Visit_ID) AS Total_Visits
FROM Fact_Mental_Health_Visit f
JOIN Dim_Time t ON f.Time_ID = t.Time_ID
GROUP BY t.Year, t.Month
ORDER BY t.Year, MONTH(STR_TO_DATE(t.Month,'%M'));
"""
df_trend = pd.read_sql(query_trend, conn)
df_trend.to_csv(os.path.join(output_folder, "Monthly_Trend.csv"), index=False)
df_trend['Month_Year'] = pd.to_datetime(df_trend['Month_Year'])

plt.figure(figsize=(10,5))
sns.lineplot(data=df_trend, x='Month_Year', y='Total_Visits', marker='o')
plt.title("Monthly Trend of Visits")
plt.ylabel("Total Visits")
plt.xlabel("Month-Year")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "Monthly_Trend.png"))
plt.close()

# -----------------------------
# 3 AI: Forecast Next 12 Months
# -----------------------------
df_trend.set_index('Month_Year', inplace=True)
model = ARIMA(df_trend['Total_Visits'], order=(1,1,1))
model_fit = model.fit()
forecast = model_fit.forecast(steps=12)

forecast_dates = pd.date_range(start=df_trend.index[-1] + pd.DateOffset(months=1), periods=12, freq='MS')
df_forecast = pd.DataFrame({'Month_Year': forecast_dates, 'Forecast_Visits': forecast})
df_forecast.to_csv(os.path.join(output_folder, "Forecast_Visits.csv"), index=False)

plt.figure(figsize=(10,5))
plt.plot(df_trend.index, df_trend['Total_Visits'], label='Historical')
plt.plot(df_forecast['Month_Year'], df_forecast['Forecast_Visits'], color='red', label='Forecast')
plt.title("Forecast of Monthly Mental Health Visits")
plt.ylabel("Total Visits")
plt.xlabel("Month-Year")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "Forecast_Visits.png"))
plt.close()

# -----------------------------
# 4 AI: High-Risk Conditions Heatmap
# -----------------------------
query_risk = """
SELECT c.Condition_Name, Severity_Level, COUNT(*) AS Visits
FROM Fact_Mental_Health_Visit f
JOIN Dim_Condition c ON f.Condition_ID = c.Condition_ID
GROUP BY c.Condition_Name, Severity_Level
ORDER BY Visits DESC;
"""
df_risk = pd.read_sql(query_risk, conn)
df_risk.to_csv(os.path.join(output_folder, "High_Risk_Conditions.csv"), index=False)

df_heatmap = df_risk.pivot(index='Condition_Name', columns='Severity_Level', values='Visits').fillna(0)

plt.figure(figsize=(10,6))
sns.heatmap(df_heatmap, annot=True, fmt='g', cmap='Reds')
plt.title("High-Risk Mental Health Conditions")
plt.ylabel("Condition")
plt.xlabel("Severity Level")
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "High_Risk_Conditions.png"))
plt.close()

# -----------------------------
# 5 Finish
# -----------------------------
conn.close()
print(f" Workflow completed! All CSVs and charts saved in folder: {output_folder}")
```
